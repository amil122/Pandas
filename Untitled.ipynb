```python
import pandas as pd

# Example DataFrame with 'Survived' as a column
data = {
    'Survived': [1, 0, 1, 0],
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [22, 28, 24, 32]
}
df1 = pd.DataFrame(data)

# Accessing the 'Survived' column
print(df1['Survived'])
print(df1.loc[:, 'Survived'])

# Example DataFrame with 'Survived' as an index
data_with_index = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [22, 28, 24, 32]
}
df_with_index = pd.DataFrame(data_with_index, index=['Survived', 'Not Survived', 'Survived', 'Not Survived'])

# Accessing the row with index 'Survived'
print(df_with_index.loc['Survived'])
```
